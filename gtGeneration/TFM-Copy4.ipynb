```python
# Función para area principal
def area_principal(img):
    gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    ret,thresh1 = cv2.threshold(gray,150,255,cv2.THRESH_BINARY)
    im2, contours, hierarchy = cv2.findContours(thresh1,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    # Guardando información de área, como coordenadas del centro, altura, ancho y angulo
    info=[]
    for c in contours:
        info.append(cv2.minAreaRect(c)) 
    # CalcuLando y guardando tamano área 
    x=[]
    for c in info:
        x.append(c[1][0]*c[1][1])
    # obteniendo coordenadas de area 
    box = np.int0(cv2.boxPoints(info[x.index(max(x))])) 

    xmax, ymax = box.max(axis=0)
    xmin, ymin = box.min(axis=0)
    area=img[ymin:ymax,xmin:xmax]

    RGB_img = cv2.cvtColor(area, cv2.COLOR_BGR2RGB)
    img_rotated = ndimage.rotate(RGB_img, info[x.index(max(x))][2])
    return  img_rotated, box
    #return  RGB_img, box
```
